```python
#faster than list demo 
import numpy as np
import time

# Python List
list_data = list(range(1_000_000))
start = time.time()
list_result = [x * 2 for x in list_data]
end = time.time()
print("Python List Time:", end - start)

# NumPy Array
np_data = np.arange(1_000_000)
start = time.time()
np_result = np_data * 2
end = time.time()
print("NumPy Time:", end - start)
```

# chapter 02 arrays, inbuilt methods, array attributes 

```python
# 1D array 
import numpy as np

a = np.array([1, 2, 3, 4, 5])
print(a)
```

```python
#1d array from python list 
list1=[1,2,3,4,5,6]
print(list1)
print(type(list1))
a=np.array(list1)
print(a)
print(type(a))
```

```python
#1d array from python tuple
tuple1=(1,2,3,4,5,6)
print(tuple1)
print(type(tuple1))
a=np.array(tuple1)
print(a)
print(type(a))
```

```python
#1d arrays from set 
set1={1,2,3,4,5,6}
print(set1)
print(type(set1))
b=np.array(set1)
print(b)
print(type(b))
```

```python
#1d array from dict 
dict1={
    "age":35,
    "salary":15000,
    "phone num":9786907526
}

print(dict1)
print(type(dict1))
b=np.array(dict1)
print(b)
print(type(b))
```
